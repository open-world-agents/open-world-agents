# Counter-Strike Deathmatch Dataset: Detailed EDA

This notebook provides a detailed exploratory data analysis (EDA) of the Counter-Strike Deathmatch dataset, including overall statistics and in-depth inspection of file structures for both `.hdf5` and `.npy` files.

```python
import os
import h5py
import numpy as np


base = "/mnt/raid12/datasets/CounterStrike_Deathmatch"
subsets = {
    "dm_expert_dust2": "dataset_dm_expert_dust2",
    "aim_expert": "dataset_aim_expert",
    "dm_expert_othermaps": "dataset_dm_expert_othermaps",
    "metadata": "dataset_metadata",
}


def get_file_stats(path, exts=None):
    if not os.path.exists(path):
        return {"count": 0, "size": 0, "files": []}
    files = [f for f in os.listdir(path) if (not exts or any(f.endswith(e) for e in exts))]
    sizes = [os.path.getsize(os.path.join(path, f)) for f in files]
    return {
        "count": len(files),
        "size": sum(sizes),
        "files": files,
        "sizes": sizes,
    }


def sizeof_fmt(num, suffix="B"):
    for unit in ["", "K", "M", "G", "T"]:
        if abs(num) < 1024.0:
            return f"{num:3.1f}{unit}{suffix}"
        num /= 1024.0
    return f"{num:.1f}P{suffix}"


# Gather stats for each subset
stats = {}
for name, relpath in subsets.items():
    path = os.path.join(base, relpath)
    if os.path.exists(path):
        if "metadata" in name:
            exts = [".npy"]
        else:
            exts = [".hdf5"]
        stat = get_file_stats(path, exts)
        stats[name] = stat
        print(f"{name}: {stat['count']} files, total size {sizeof_fmt(stat['size'])}, e.g. {stat['files'][:3]}")
    else:
        print(f"{name}: path not found")
```

## Inspecting .hdf5 File Structure

For a sample `.hdf5` file from the `aim_expert` subset, we list all keys and show the shape and dtype of each dataset.

```python
# Inspect a sample .hdf5 file in aim_expert
aim_path = os.path.join(base, "dataset_aim_expert")
h5_files = [f for f in os.listdir(aim_path) if f.endswith(".hdf5")]
if h5_files:
    sample_file = os.path.join(aim_path, h5_files[0])
    print(f"Sample .hdf5 file: {sample_file}")
    with h5py.File(sample_file, "r") as f:
        keys = list(f.keys())
        print(f"Total keys: {len(keys)}")
        for k in keys:
            d = f[k]
            if hasattr(d, "shape"):
                print(f"{k}: shape {d.shape}, dtype {d.dtype}")
            else:
                print(f"{k}: type {type(d)}")
else:
    print("No .hdf5 files found in aim_expert.")
```

## Inspecting .npy Metadata File Structure

For a sample `.npy` file from the `metadata` subset, we load the file and summarize its structure: number of top-level keys, example key names, and the type/shape of a sample value.

```python
# Inspect a sample .npy metadata file
meta_path = os.path.join(base, "dataset_metadata")
npy_files = [f for f in os.listdir(meta_path) if f.endswith(".npy")]
if npy_files:
    sample_npy = os.path.join(meta_path, "currvarsv2_aim_july2021_expert_1_to_100.npy")
    print(f"Sample .npy file: {sample_npy}")
    arr = np.load(sample_npy, allow_pickle=True).item()
    print(f"Top-level keys: {len(arr)}")
    keys = list(arr.keys())
    print(f"Example keys: {keys[:3]}")
    sample_val = arr[keys[0]]
    print(f"Sample value type: {type(sample_val)}")
    if isinstance(sample_val, (list, tuple)) and len(sample_val) > 0:
        print(f"Sample value[0] type: {type(sample_val[0])}")
        if hasattr(sample_val[0], "shape"):
            print(f"Sample value[0] shape: {sample_val[0].shape}")
        if isinstance(sample_val[0], dict):
            print(f"Sample value[0] dict keys: {list(sample_val[0].keys())}")
else:
    print("No .npy files found in metadata.")
```
